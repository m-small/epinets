```julia
i=2
p1=Plots.plot(r2s,p2s,log.(mmi[i,:,:])./log(10),st=:surface,xlabel="r",ylabel="p",zlabel="log(max I)",colorbar=false,title="(a) $(Int(C[i]*100))%" )
i=4
p2=Plots.plot(r2s,p2s,log.(mmi[i,:,:])./log(10),st=:surface,xlabel="r",ylabel="p",zlabel="log(max I)",colorbar=false,title="(b) $(Int(C[i]*100))%" )
i=5
p3=Plots.plot(r2s,p2s,log.(mmi[i,:,:])./log(10),st=:surface,xlabel="r",ylabel="p",zlabel="log(max I)",colorbar=false,title="(c) $(Int(C[i]*100))%" )
i=6
p4=Plots.plot(r2s,p2s,log.(mmi[i,:,:])./log(10),st=:surface,xlabel="r",ylabel="p",zlabel="log(max I)",colorbar=false,title="(d) $(Int(C[i]*100))%" )
Plots.plot(p1,p2,p3,p4,layout=(2,2),size=(1000,800))
```

```julia
labs=["r=$x" for x in r2s]
labs=permutedims(labs)
i=3
p1=plot(p2s,mmi[i,:,:],ylabel="total",legend=:topleft,lw=2,label=labs,zlabel="log(total I)",colorbar=false,title="(a) $(Int(C[i]*100))%" )
i=4
p2=plot(p2s,mmi[i,:,:],ylabel="total infections",legend=:none,lw=2,label=labs,colorbar=false,title="(a) $(Int(C[i]*100))%" )
i=5
p3=plot(p2s,mmi[i,:,:],yaxis=:log,ylabel="total infections",legend=:topleft,lw=2,label=labs,colorbar=false,title="(b) $(Int(C[i]*100))%" )
i=6
p4=plot(p2s,mmi[i,:,:],yaxis=:log,xlabel="p",ylabel="total infections",legend=:none,lw=2,label=labs,colorbar=false,title="(c) $(Int(C[i]*100))%" )
```
